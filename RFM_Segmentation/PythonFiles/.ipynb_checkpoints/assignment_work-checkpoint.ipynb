```python
from datetime import datetime
import pandas as pd

date_diff_col = []

def calculate_date_difference(date1, date2):
    date_format = "%Y-%m-%d"  # Modify the format based on your input
    start_date = datetime.strptime(date1, date_format)
    end_date = datetime.strptime(date2, date_format)
    difference = abs((end_date - start_date).days)
    return difference

# Example usage
date2 = "2022-12-31" 

date1_list = pd.to_datetime(date1_list)  # Convert to Pandas Timestamp objects

for date1 in date1_list:
    date1_str = date1.strftime("%Y-%m-%d")  # Convert Timestamp to string
    difference_in_days = calculate_date_difference(date1_str, date2)
    date_diff_col.append(difference_in_days)
```

```python
###RFM Segmentation
```

```python
quantiles = cust_merge[['frequency','monetary_value','recency' ]].quantile(q=[0.25,0.5,0.75])
quantiles
```

```python
quantiles = quantiles.to_dict()
quantiles
```

```python
#I have created two classes for the RFM segmentation since, being high recency is bad, while high frequency and monetary value is good.
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def RClass(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FMClass(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
cust_merge['R_Quartile'] = cust_merge['recency'].apply(RClass, args=('recency',quantiles,))
cust_merge['F_Quartile'] = cust_merge['frequency'].apply(FMClass, args=('frequency',quantiles,))
cust_merge['M_Quartile'] = cust_merge['monetary_value'].apply(FMClass, args=('monetary_value',quantiles,))
```

```python
cust_merge1=cust_merge.assign(R_seg=np.nan, FM_seg =np.nan)
cust_merge1.R_seg[cust_merge['R_Quartile']==4] = "3.Churn"
cust_merge1.R_seg[cust_merge['R_Quartile']==2] = "2.At Risk"
cust_merge1.R_seg[cust_merge['R_Quartile']==3] = "2.At Risk"
cust_merge1.R_seg[cust_merge['R_Quartile']==1] = "1.Active"

cust_merge1.FM_seg[(cust_merge['F_Quartile']==3) & (cust_merge['M_Quartile']==1)] = "2.Gold"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==4) & (cust_merge['M_Quartile']==1)] = "2.Gold"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==2) & (cust_merge['M_Quartile']==2)] = "2.Gold"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==2) & (cust_merge['M_Quartile']==1)] = "2.Gold"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==3) & (cust_merge['M_Quartile']==2)] = "2.Gold"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==4) & (cust_merge['M_Quartile']==2)] = "1.Silver"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==1) & (cust_merge['M_Quartile']==3)] = "2.Gold"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==2) & (cust_merge['M_Quartile']==3)] = "2.Gold"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==3) & (cust_merge['M_Quartile']==3)] = "1.Silver"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==4) & (cust_merge['M_Quartile']==3)] = "1.Silver"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==1) & (cust_merge['M_Quartile']==4)] = "2.Gold"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==1) & (cust_merge['M_Quartile']==2)] = "2.Gold"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==2) & (cust_merge['M_Quartile']==4)] = "1.Silver"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==3) & (cust_merge['M_Quartile']==4)] = "1.Silver"
cust_merge1.FM_seg[(cust_merge['F_Quartile']==4) & (cust_merge['M_Quartile']==4)] = "0.Standard"

cust_merge1.FM_seg.fillna("4.Premium", inplace=True)
```
